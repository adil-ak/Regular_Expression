**findall()**

```python
import re
sentence="Select words and 123 to get definitions and 456 translations without leaving the page"
f=re.findall("\d+",sentence)
print(f)
```

\d is used to get the digits   
\d+ - used to get digits which is together

```python
import re
sentence="Select words and 123 to get definitions and 456 translations without leaving the page"
f=re.findall("\d",sentence)
print(f)
```

**\D - Capital letter D used to get non digit words including spaces**

```python
import re
sentence="Select words and 123 to get definitions and 456 translations without leaving the page"
f=re.findall("\D",sentence)
print(f)
```

**\w - Use to get characters without special characters**

```python
sent="hiwith123#$%&"
t=re.findall("\w",sent)
print(t)
```

**\W - Capital W to get the special characters**

```python
y=re.findall("\W",sent)
print(y)
```

**\s -  Used to get the count of Spaces**





```python
sent="hi adil im su su suresh 123"
s=re.findall("\s",sent)
print(s)
```

```python
sent="hi adil im su su suresh 123"
s=re.findall("\s",sent)
len(s)
```

```python
h=re.findall("\s","Cranes Varsity Private limited")
len(h)
```

```python
import re
sentence="india is country.country is good country"
pattern="country"
res=re.findall(pattern,sentence)
if len(res)==0:
  print(" not found")
else:
  print(f"found {pattern}",len(res)," times")
```



> finditer()



```python
import re
sentence="india is country.country is good country"
pattern="country"
res=re.finditer(pattern,sentence)
for i in res:
  print(i)
```

**fullmatch()**

```python
import re
sentence="Anoop"
pattern="Anoop"
res=re.fullmatch(pattern,sentence)
print(res)
```
